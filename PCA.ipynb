```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import load_iris
```

```python
iris = load_iris()
iris
```

```python
x=iris.data
y=iris.target
```

```python
y
```

```python
#X作为数组，是二维
x.shape
```

```python
import pandas as pd
#作为表或矩阵，x是4维（4个特征）  
pd.DataFrame(x)
```

```python
from sklearn import decomposition

pca = decomposition.PCA(n_components=2)
pca = pca.fit(x)
X_dr = pca.transform(x) #获取降维后的新矩阵
```

```python
X_dr.shape  #二维
```

```python
X_dr
```

```python
s=X_dr[y==0]
s
```

```python
s.shape
```

```python
plt.figure()
plt.scatter(X_dr[y==0,0],X_dr[y==0,1],color='red',label=iris.target_names[0])  #y==0,y==1是布尔索引
plt.scatter(X_dr[y==1,0],X_dr[y==1,1],color='blue',label=iris.target_names[1])
plt.scatter(X_dr[y==2,0],X_dr[y==2,1],color='green',label=iris.target_names[2])
plt.legend()    #对图中点作标注
plt.title('PCA of iris datasets')
plt.show()
```

```python
#以上代码可以用for 循环执行
color =['red','blue','green']
plt.figure(figsize=(10,5))
for i in [0,1,2]:
    plt.scatter(X_dr[y==i,0],X_dr[y==i,1]
                ,color=color[i]
                ,label=iris.target_names[i] 
                ,alpha=.6  #点的透明度
               )
plt.legend()    #标注
plt.title('PCA of iris datasets')
plt.show()
    
```

```python
pca.explained_variance_     #降维后的新特征向量上所带信息量的大小(可解释性方差的大小)
```

```python
pca.explained_variance_ratio_  #新特征向量所带信息占原始数据总信息量的百分比
```

```python
pca.explained_variance_ratio_ .sum()
```

```python
#n_components不赋值默认是min(data.shape),一般是特征个数
pca_line = decomposition.PCA().fit(x)  #实例化，拟合，获取  一步到位;
```

```python
pca_line.explained_variance_ratio_
```

```python
np.cumsum(pca_line.explained_variance_ratio_)   #cumsum()是numpy数组中前i个数相加
```

```python
#如何选取n_components的整数大小，可以通过可解释性方差贡献率曲线显示
from pylab import * 
mpl.rcParams['font.sans-serif'] = ['SimHei'] #在坐标轴上指定默认字体  （中文不乱码）
plt.plot([1,2,3,4],np.cumsum(pca_line.explained_variance_ratio_))
#plt.xticks([1,2,3,4])
plt.xlabel("the value of n_components 整数")
plt.ylabel("explained_variance_ratio")
plt.show()
```

```python
pca_mle=decomposition.PCA(n_components="mle")  #mle：最大似然估计作为自选超参数
pca_mle=pca_mle.fit(x)
X_mle=pca_mle.transform(x)
```

```python
pca_f = decomposition.PCA(n_components=0.96,svd_solver='full') #  0<n_components<1 的浮点数，自动选出信息量超过该浮点数的特征数量,svd_solver分解器
pca_f = pca_f.fit(x)
X_f = pca_f.transform(x)
```

```python
pca= decomposition.PCA(150)
pca=pca.fit(x)
x_sd=pca.transform(x)
```

```python
v=pca.components_    #V(k,n)
pca.n_components_    #降维后的特征维度
```

```python
v[0].reshape(62,47).shape
```

```python
#查看新空间向量的特征提取
fig,axes = plt.subplots(4,5,
                       figsize=(8,8)
                      ,subplot_kw={"xticks":[],"yticks":[]}
                       )
#axes[0][0].imshow(y[0,:,:])   一张图
for i,ax in enumerate(axes.flat):
    ax.imshow(v[i,:].reshape(62,47),cmap="gray")
```

```python
x_inverse=pca.inverse_transform(x_sd)  #inverse_transform降维后的空间逆转到原始的空间维度，但并不能完全还原
x_inverse
```

```python
x_inverse.shape  #还原成原始的向量空间
```
